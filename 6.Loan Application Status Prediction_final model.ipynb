# Loan Application Status Prediction

# 1. Introduction

# Project Description
This dataset includes details of applicants who have applied for loan. The dataset includes details like credit history, loan amount, their income, dependents etc.

In this project, we'll develop a machine learning model to predict whether loan applications will be approved. We'll use applicant details like demographics, financial status, and credit history to train the model. The aim is to automate and optimize the loan approval process, making it more efficient and accurate for financial institutions. Key steps include data preprocessing, feature engineering, model selection, training, and evaluation. Ethical considerations regarding fairness and privacy will also be addressed. Ultimately, the goal is to improve lending practices and financial outcomes.

# Dataset Attributes
Certainly! Here's a concise summary of the dataset attributes:

1. Loan_ID: Unique identifier for the applicant's affirmed purchases.
2. Gender: Categorization of applicants into male and female based on reproductive functions.
3. Married: Indicates whether the applicant is in a state of matrimony.
4. Dependents: Persons who depend on the applicant for survival.
5. Education: Number of years the applicant received systematic instruction, particularly at a school or university.
6. Self_Employed: Indicates whether the applicant is self-employed or works for an employer.
7. Applicant Income: Disposable income available for the applicant's use under State law.
8. CoapplicantIncome: Disposable income available for individuals participating in the loan application alongside the main applicant, as per State law.
9. Loan_Amount: The amount of money the applicant owes at any given time.
10. Loan_Amount_Term: Duration for which the loan is availed to the applicant.
11. Credit History: Record of the applicant's ability to repay debts and demonstrated responsibility in repaying them.
12. Property_Area: Total area within the boundaries of the property as set out in the schedule.
13. Loan_Status: Indicates whether the applicant is eligible to be availed the requested loan.

# Objective:
The objective of this project is to build a machine learning model that predicts whether a loan application will be approved or not based on various applicant details such as credit history, income, education, and property area. By accurately predicting loan approval status, financial institutions can streamline their lending processes, minimize risks, and improve decision-making.

# Output
The output of the loan application status prediction model will be a binary classification indicating whether the loan application is approved or not.

# Approach_Outlier-II:-Removing outliers using the IQR method.

```python
# Convert all columns to numeric types
df_numeric = df.apply(pd.to_numeric, errors='coerce')



# Calculate quartiles
q1 = df_numeric.quantile(0.25)
q3 = df_numeric.quantile(0.75)

# Calculate interquartile range (IQR)
IQR = q3 - q1

# Calculate upper and lower bounds for outliers
upper_bound = q3 + (1.5 * IQR)
lower_bound = q1 - (1.5 * IQR)
```

```python
ApplicantIncome_high=(q3.ApplicantIncome+(1.5*IQR.ApplicantIncome))


index=np.where(df['ApplicantIncome']> ApplicantIncome_high)

df=df.drop(df.index[index])
```

```python
LoanAmount_high=(q3.LoanAmount+(1.5*IQR.LoanAmount))


index=np.where(df['LoanAmount']>LoanAmount_high)

df=df.drop(df.index[index])
```

```python
ApplicantIncome_low=q1.ApplicantIncome-(1.5*IQR.ApplicantIncome)

index=np.where(df['ApplicantIncome']<ApplicantIncome_low)

df=df.drop(df.index[index])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import  r2_score
maxacc=0
maxrs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=i) 
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)   
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('Best accuracy is', maxacc,'at random state',maxrs) 
```
